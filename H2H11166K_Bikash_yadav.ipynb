# Milestone 4 - ML Models and Evaluations

```python
df.head(15)
```

```python
df.fillna(0,inplace=True)
df.head(2)
print(df.info())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
MSE_Score = []
R2_Score = []
Algorithm = []
```

```python
feat_list=['PURCHASE_ORDER_TYPE',
'RELEASED_CREDIT_VALUE',
'ORDER_CREATION_TIME',
'ORDER_CURRENCY',
'CUSTOMER_NUMBER',
'Last-1day_Sales',
'Last-2day_Sales',
'Last-3day_Sales',
'Last-4day_Sales',
'Last-5day_Sales',
'Last-6day_Sales',
'Last-7day_Sales'
]
targ_list=['amount_in_usd'] #


X = df[feat_list]
y = df[targ_list]

# # Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### LINEAR REGRESSION

```python
# Initialize and train the Linear Regression model
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = linear_regression.predict(X_test)

# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted))
R2_Score.append(r2_score(y_test, predicted))
Algorithm.append('LinearRegression')
```

#### Decision Tree Regression

```python
# Fitting Decision Tree to the Training Set
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = reg.predict(X_test)

# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted))
R2_Score.append(r2_score(y_test, predicted))
Algorithm.append('DecisionTreeRegression')
```

#### RANDOM FOREST


```python
#import RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,y_train)

# Predicting the Test Set Results
predicted=rf.predict(X_test)

# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted))
R2_Score.append(r2_score(y_test, predicted))
Algorithm.append('RandomForestRegression')
```

#### ADABOOST

```python
from sklearn.ensemble import AdaBoostRegressor
ab = AdaBoostRegressor()
ab.fit(X_train, y_train)

# Predicting the Test Set Results
predicted=ab.predict(X_test)

# Appending the Scores For Visualisation at a Later Part
MSE_Score.append(mean_squared_error(y_test, predicted))
R2_Score.append(r2_score(y_test, predicted))
Algorithm.append('AdaBoostRegression')
```

#### XGBOOST

```python
import xgboost as xgb
# Fitting XGBoost Regressor to the Training Set
reg = xgb.XGBRegressor()
reg.fit(X_train, y_train)

# Predicting the Test Set Results
predicted = reg.predict(X_test)

# Appending the Scores For Visualisation at a Later Part
Algorithm.append('Extreme Gradient Boost Algorithm')
MSE_Score.append(mean_squared_error(y_test, predicted))
R2_Score.append(r2_score(y_test, predicted))
```
